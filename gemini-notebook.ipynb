## Principle 1: Clear & Specific Instructions
From the DeepLearning.AI course — vague prompts give vague answers.
Demonstrating the difference with an interview question example.

````python
model = genai.GenerativeModel("gemini-2.5-flash")

# Vague prompt — bad
vague = model.generate_content("Give me an interview question")
print("VAGUE:", vague.text)

# Specific prompt — good (what the course taught)
specific = model.generate_content("""
Generate ONE technical interview question for a
Senior React developer role focused on performance
optimization. Format: Question only, no answer.
""")
print("SPECIFIC:", specific.text)
````

````python
import json
import re


def parse_json_from_text(text: str):
    """Parse JSON robustly from model text (raw, fenced, or mixed output)."""
    if text is None:
        raise ValueError("Model returned None text")

    cleaned = text.strip()
    if not cleaned:
        raise ValueError("Model returned empty text")

    # Remove markdown code fences if present.
    cleaned = re.sub(r"^```(?:json)?\s*", "", cleaned, flags=re.IGNORECASE)
    cleaned = re.sub(r"\s*```$", "", cleaned)

    # Try direct parse first.
    try:
        return json.loads(cleaned)
    except json.JSONDecodeError:
        pass

    # Fallback: extract first JSON object from text.
    match = re.search(r"\{[\s\S]*\}", cleaned)
    if not match:
        raise ValueError(f"No JSON object found in response: {text[:200]}")

    return json.loads(match.group(0))


score_prompt = """
Based on this interview, return ONLY a JSON object with this exact schema:
{
  "score": <integer 1-10>,
  "verdict": "<one short verdict>",
  "strengths": ["<item>", "<item>"],
  "improvements": ["<item>", "<item>"]
}
No extra text. JSON only.
"""

# Ask Gemini for strict JSON output.
result = chat.send_message(
    score_prompt,
    generation_config={"response_mime_type": "application/json"}
)

parsed = parse_json_from_text(result.text)
print(f"Score: {parsed['score']}/10")
print(f"Verdict: {parsed['verdict']}")
print("Strengths:", parsed["strengths"])
print("Improvements:", parsed["improvements"])
````
